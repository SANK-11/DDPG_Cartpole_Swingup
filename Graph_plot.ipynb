```python
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_learning_curve_ddpg(log_folder="logs"):
    files = glob.glob(os.path.join(log_folder, "*.monitor.csv"))
    
    data_frames = []

    print(f"Found {len(files)} log files. Processing...")

    for file in files:
        df = pd.read_csv(file, skiprows=1)
            
        df.columns = [c.strip() for c in df.columns]
        
        if 'r' not in df.columns or 'l' not in df.columns:
            print(f"Skipping {file}: 'r' or 'l' column missing.")
            continue

       
        df['Timesteps'] = df['l'].cumsum()
        df['Reward'] = df['r']
        
        data_frames.append(df)

    if len(data_frames) == 0:
        print("No valid log files found. Exiting.")
        return

    full_df = pd.concat(data_frames, ignore_index=True)

    plt.figure(figsize=(10, 6))
    sns.set_theme(style="darkgrid")

    sns.lineplot(
        data=full_df, 
        x='Timesteps', 
        y='Reward', 
        errorbar='sd'  
    )

    plt.title("DDPG on Cartpole-Swingup (Mean Â± Std over 3 Seeds)", fontsize=14)
    plt.xlabel("Total Timesteps", fontsize=12)
    plt.ylabel("Episode Reward", fontsize=12)
    plt.axhline(y=800, color='r', linestyle='--', label='High Performance Threshold')
    plt.legend()

    plt.tight_layout()

    output_filename = "ddpg_evalution_check.png"
    plt.savefig(output_filename, dpi=500)
    print(f"Plot saved as {output_filename}")
    
    plt.show()
```

```python
plot_learning_curve_ddpg()
```
